```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

# Data penjualan minggu 1 sampai 5
x = [[1], [2], [3], [4], [5]]  # x adalah minggu, sifat : independent
y = [[150], [157], [162], [166], [177]]  # y adalah data penjualan di minggu tersebut. sifat : dependent

# Training data
x_train = x
y_train = y

# Membuat model regresi linear
lm = LinearRegression()
lm.fit(x_train, y_train)  # Fase training

# Prediksi penjualan untuk minggu ke-6
x_test = [[6]]  # Data yang akan diprediksi (minggu ke-6)
p = lm.predict(x_test)  # Fase prediksi
print('Angka ramalan untuk penjualan minggu ke-6 :', p[0][0])

# Plot hasil prediksi
pb = lm.predict(x_train)
plt.scatter(x_train, y_train, label="Data penjualan")
plt.plot(x_train, pb, color='red', linewidth=2, label="Garis prediksi")
plt.xlabel('Minggu')
plt.ylabel('Penjualan')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Data Bulan dan Penjualan
bulan = [[1], [2], [3], [4], [5]]  # X: bulan
penjualan = [[220], [250], [210], [300], [325]]  # Y: penjualan

# Membuat model regresi linier
model = LinearRegression()
model.fit(bulan, penjualan)  # Fase training

# Mendapatkan nilai slope (b) dan intercept (a)
b = model.coef_[0][0]
a = model.intercept_[0]
print(f"Persamaan regresi: Y = {a:.2f} + {b:.2f}X")

# Peramalan untuk bulan 1 hingga 5
peramalan = model.predict(bulan)

# Menampilkan tabel peramalan
df = pd.DataFrame({
    'Bulan': [1, 2, 3, 4, 5],
    'Penjualan Aktual': [220, 250, 210, 300, 325],
    'Penjualan Ramalan': peramalan.flatten().round(2)
})
print(df)

# Plot hasil peramalan
plt.scatter(bulan, penjualan, color='blue', label='Aktual')
plt.plot(bulan, peramalan, color='red', label='Ramalan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.show()
```

```python
# Menghitung MAPE
aktual = np.array([220, 250, 210, 300, 325])
ramalan = peramalan.flatten()
mape = np.mean(np.abs((aktual - ramalan) / aktual)) * 100
print(f"MAPE: {mape:.2f}%")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor

# Data Bulan dan Penjualan
bulan = [[1], [2], [3], [4], [5]]  # X: bulan
penjualan = [220, 250, 210, 300, 325]  # Y: penjualan

# Membuat model Random Forest Regressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(bulan, penjualan)  # Fase training

# Peramalan untuk bulan 1 hingga 5
peramalan = model.predict(bulan)

# Menampilkan tabel peramalan
df = pd.DataFrame({
    'Bulan': [1, 2, 3, 4, 5],
    'Penjualan Aktual': [220, 250, 210, 300, 325],
    'Penjualan Ramalan': peramalan.round(2)
})
print(df)

# Menghitung MAPE
mape = np.mean(np.abs((df['Penjualan Aktual'] - df['Penjualan Ramalan']) / df['Penjualan Aktual'])) * 100
print(f'MAPE: {mape:.2f}%')

# Plot hasil peramalan
plt.scatter(bulan, penjualan, color='blue', label='Aktual')
plt.plot(bulan, peramalan, color='green', label='Garis Regresi')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.show()
```
