```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import table

# Membaca data dari file Excel
df = pd.read_excel('Sample_Superstore.xlsx')

# Mengelompokkan berdasarkan 'Customer Name' dan menjumlahkan profit
top_customers = df[['Customer Name', 'Profit']].groupby('Customer Name').sum()

# Mengurutkan berdasarkan profit terbanyak dan memilih 10 pelanggan teratas
top_customers = top_customers.sort_values('Profit', ascending=False).head(10)

# Pembulatan nilai profit menjadi 2 desimal
top_customers = top_customers[["Profit"]].round(2)

# Mereset index agar 'Customer Name' kembali menjadi kolom
top_customers.reset_index(inplace=True)

# Membuat grafik
fig, ax = plt.subplots(figsize=(15, 10))

# Membuat bar chart horizontal
ax.barh(top_customers['Customer Name'], top_customers['Profit'], color='skyblue', edgecolor='black', linewidth=1)

# Memberikan label pada sumbu x dan y
ax.set_xlabel('Profit ($)', fontsize=15)
ax.set_ylabel('Nama Pelanggan', fontsize=15)
ax.set_title('10 Pelanggan dengan Profit Terbanyak', fontsize=18)

# Menambahkan nilai profit di samping setiap batang
for i in range(top_customers.shape[0]):
    ax.text(top_customers['Profit'].iloc[i] + 500, i, '$'+str(top_customers['Profit'].iloc[i]), fontsize=12, color='black')

# Menambahkan tabel di atas grafik
ax_table = plt.gca().inset_axes([0, 0.7, 1, 0.2])  # Menambahkan axes untuk tabel
table_data = top_customers.copy()
table_data.columns = ['Customer Name', 'Profit ($)']  # Menyusun ulang kolom tabel
tbl = table(ax_table, table_data, loc='top', cellLoc='center', colWidths=[0.2, 0.2])
tbl.auto_set_font_size(False)
tbl.set_fontsize(12)
tbl.scale(1.2, 1.2)  # Menyesuaikan ukuran tabel

# Menyembunyikan sumbu tabel
ax_table.axis('off')

# Menampilkan grafik dan tabel
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

newdf=df.copy()

newdf.set_index('Order Date', inplace = True)

monthly_sales = newdf['Sales'].resample('ME').sum()

plt.figure(figsize=(12,6))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values)
plt.title('Monthly Sales Trend - Sample Superstore')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca data dari file Excel
df = pd.read_excel('Sample_Superstore.xlsx')

# Mengubah kolom 'Order Date' menjadi datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Mengubah kolom 'Sales' menjadi numerik (jika belum)
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')

# Menetapkan 'Order Date' sebagai indeks
df.set_index('Order Date', inplace=True)

# Menghitung total penjualan per tahun
years_sales = df.resample('Y').sum()['Sales']

# Membuat grafik
plt.figure(figsize=(10, 5))
sns.barplot(x=years_sales.index.year, y=years_sales.values, palette='viridis')
plt.title('Years Sales Trend - Sample Superstore')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.xticks(rotation=45)  # Memutar label tahun jika perlu
plt.show()
```
